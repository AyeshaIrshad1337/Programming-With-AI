```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("winequality-red.csv")
df.head()
```

```python
f=df["fixed acidity"]
s=df["sulphates"]
plt.scatter(f,s)
plt.show()
```

```python
free_sulfur_dioxide = df['free sulfur dioxide']
total_sulfur_dioxide = df['total sulfur dioxide']
alcohol = df['alcohol']
fig, ax = plt.subplots()
ax.boxplot(free_sulfur_dioxide)
plt.show()
```

```python
columns=[f,free_sulfur_dioxide, total_sulfur_dioxide, alcohol]
fig, ax = plt.subplots()
ax.boxplot(columns)
plt.show()
```
