```python
import pandas as pd
df=pd.read_csv("/kaggle/input/diabetes-dataset/diabetes.csv")
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates()
df.dropna()
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
array=df.values
X=array[:,0:8]
Y=array[:,8]
t_size=0.20
seed=4
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=t_size,random_state=seed)
```

```python
#Decision Tree using Entropy
dt=DecisionTreeClassifier(criterion='entropy',ccp_alpha=0.015)
dt.fit(X_train,Y_train)
prediction=dt.predict(X_test)
print((accuracy_score(Y_test,prediction)))
```

```python
dt = DecisionTreeClassifier( criterion='gini', ccp_alpha = 0.015) #ccp_alpha is pruning parameter   
dt.fit(X_train, Y_train)      # training the model
predictions = dt.predict(X_test)  # now testing on test data to get class of test data
print((accuracy_score(Y_test, predictions))) # comparing results predicted by model with actual to get accuracy score
```

```python
from sklearn.naive_bayes import GaussianNB

# Naive Bayes Algorithm
nb = GaussianNB() #ccp_alpha is pruning parameter   
nb.fit(X_train, Y_train)      # training the model
predictions = nb.predict(X_test)  # now testing on test data to get class of test data
print((accuracy_score(Y_test, predictions))) # comparing results predicted by model with actual to get accuracy score
```
