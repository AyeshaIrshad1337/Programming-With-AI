```python
import pandas as pd
data=pd.read_csv("/kaggle/input/data-for-admission-in-the-university/adm_data.csv")
data.head()
```

```python
data.duplicated().sum()
X = data[['GRE Score', 'TOEFL Score','CGPA','Research']]
Y = data[["Chance of Admit "]]
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, Y)
```

```python
predictedChances = regr.predict([[230,105,8.9,1]])
print(predictedChances)
```
